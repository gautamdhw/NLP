<h2>MARKOV CHAIN FUNCTION</h2>

```python
from collections import defaultdict

def markov_chain(text):
    '''The input is a string of text and the output will be a dictionary with each word as
       a key and each value as the list of words that come after the key in the text.'''

    # Tokenize the text by word, though including punctuation
    words = text.split(' ')

    # Initialize a default dictionary to hold all of the words and next words
    m_dict = defaultdict(list)

    # Create a zipped list of all of the word pairs and put them in word: list of next words format
    for current_word, next_word in zip(words[0:-1], words[1:]):
        m_dict[current_word].append(next_word)

    # Convert the default dict back into a dictionary
    m_dict = dict(m_dict)
    return m_dict
```

<h2>TEXT GENERATOR</h2>

```python
import random

def generate_sentence(chain, count=15):
    '''Input a dictionary in the format of key = current word, value = list of next words
       along with the number of words you would like to see in your generated sentence.'''

    # Capitalize the first word
    word1 = random.choice(list(chain.keys()))
    sentence = word1.capitalize()

    # Generate the second word from the value list. Set the new word as the first word. Repeat.
    for i in range(count-1):
        word2 = random.choice(chain[word1])
        word1 = word2
        sentence += ' ' + word2

    # End it with a period
    sentence += '.'
    return(sentence)
```

<h2>IMPROVED TEXT GENERATOR</h2>

- Ends the sentence early if a word ends with `.`, `!`, or `?`  
- Prevents errors by checking if the next word exists in the chain  
- Adds a random punctuation mark at the end if needed

```python
import random
import string

def generate_sentence1(chain, count=15):
    '''Generate a sentence using a Markov chain.
       Ends early if a word ends with punctuation, or finishes with random punctuation.'''

    # Choose a random starting word
    word1 = random.choice(list(chain.keys()))
    sentence = word1.capitalize()

    for _ in range(count - 1):
        # Agar word1 ka last character ., !, ya ? hai — matlab sentence khatam ho chuka hai — toh loop se bahar nikal jao.
        if word1[-1] in '.!?':
            break

        # Agar current word chain me hai hi nahi ya uska koi next word nahi hai, toh bhi loop khatam.
        if word1 not in chain or not chain[word1]:
            break

        word2 = random.choice(chain[word1])
        sentence += ' ' + word2
        word1 = word2

    # Agar sentence abhi tak punctuation mark se end nahi ho raha hai, toh randomly ., !, ya ? add kar dete hain.
    if sentence[-1] not in '.!?':
        sentence += random.choice(['.', '!', '?'])

    return sentence
```

## HOW MARKOV IS WORKING

```python
def markov_chain_verbose(text):
    '''Markov chain with full calculation printed step-by-step.'''
    
    words = text.split(' ')
    print("Words List:", words)
    
    m_dict = defaultdict(list)
    
    print("\nMaking Word Pairs and Updating Dictionary:\n")
    for current_word, next_word in zip(words[0:-1], words[1:]):
        print(f"Current Word: '{current_word}' -> Next Word: '{next_word}'")
        m_dict[current_word].append(next_word)
        print(f"Updated Dictionary so far: {dict(m_dict)}\n")
    
    m_dict = dict(m_dict)
    print("\nFinal Markov Chain Dictionary:")
    print(m_dict)
    
    return m_dict
```

```python
verbose_dict=markov_chain_verbose('I like pizza and i like pasta')
```

```python
def generate_sentence_verbose(chain, count=20):
    '''Input a Markov chain and generate a sentence with detailed step-by-step explanation.'''
    
    # Start with a random first word
    word1 = random.choice(list(chain.keys()))
    sentence = word1.capitalize()
    print(f"Starting word: '{word1}'\nSentence till now: {sentence}\n")
    
    # Generate next words
    for i in range(count-1):
        possible_next_words = chain.get(word1, [])
        
        if not possible_next_words:
            print(f"No next words found for '{word1}'. Stopping early.")
            break
        
        word2 = random.choice(possible_next_words)
        sentence += ' ' + word2
        print(f"Step {i+1}:")
        print(f"Current Word: '{word1}' -> Chose Next Word: '{word2}'")
        print(f"Sentence till now: {sentence}\n")
        
        word1 = word2  # Update the current word to the new word
    
    # End with a period
    sentence += '.'
    print(f"Final Sentence: {sentence}")
    return sentence
```

```python
generate_sentence_verbose(verbose_dict)
```
